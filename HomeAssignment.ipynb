```python
from matplotlib import pyplot

# get the data from the functional data file
func_img = image.load_img(all_files[1])
data = func_img.get_fdata()

# obtain only positive values
positive_values = data[data > 0]

# plot the histogram
fig, ax = plt.subplots()
ax.hist(positive_values, color = 'tab:pink')

# set title
ax.set_title('Voxel Activation Distribution of Meta-Analysis (N=79)')
ax.set_ylabel('Number of Voxels')
ax.set_xlabel('Z-Score')
```
